# Random Forest Regressor

```python
import os
import re
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

from audio_related import extract_features


dataset_path = r"C:\Users\jashw\OneDrive\Desktop\Sem 4\Robotics\Project\train\thickness\data\Mild Stell"

X = []
Y = []

print("Loading dataset...!")

for folder in os.listdir(dataset_path):

    folder_path = os.path.join(dataset_path, folder)

    if not os.path.isdir(folder_path):
        continue

    thickness = float(re.findall(r"\d+\.?\d*", folder)[0])

    for file in os.listdir(folder_path):

        if not file.endswith(".wav"):
            continue

        file_path = os.path.join(folder_path, file)

        try:
            features = extract_features(file_path)

            X.append(features)
            Y.append(thickness)

        except Exception as e:
            print("Error:", file_path, e)

X = np.array(X)
Y = np.array(Y)

print("Total samples:", len(X))
# print("Thickness values:", float(np.unique(Y)))
print("Thickness values:", np.unique(Y))

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=20
)

model = RandomForestRegressor(
    n_estimators=500,
    random_state=30
)

print("Training model...!")
model.fit(X_train, Y_train)


predictions = model.predict(X_test)

mae = mean_absolute_error(Y_test, predictions)

print("\nRESULTS")
print("MAE:", mae)
print("Sample Predictions:", predictions[:10])

joblib.dump(model, "thickness_model_rfr_ms.pkl")
joblib.dump(scaler, "scaler_rfr_ms.pkl")

print("\nModel saved successfully.")
```

# SVM Regressor

```python
import os
import re
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR

from audio_related import extract_features


dataset_path = r"C:\Users\jashw\OneDrive\Desktop\Sem 4\Robotics\Project\train\thickness\data\Mild Stell"

X = []
Y = []

print("Loading dataset...!")

for folder in os.listdir(dataset_path):

    folder_path = os.path.join(dataset_path, folder)

    if not os.path.isdir(folder_path):
        continue

    thickness = float(re.findall(r"\d+\.?\d*", folder)[0])

    for file in os.listdir(folder_path):

        if not file.endswith(".wav"):
            continue

        file_path = os.path.join(folder_path, file)

        try:
            features = extract_features(file_path)

            X.append(features)
            Y.append(thickness)

        except Exception as e:
            print("Error:", file_path, e)

X = np.array(X)
Y = np.array(Y)

print("Total samples:", len(X))
# print("Thickness values:", float(np.unique(Y)))
print("Thickness values:", np.unique(Y))

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=20
)

model = SVR(
    kernel="rbf",
    C=100,
    gamma="scale"
)

print("Training model...!")
model.fit(X_train, Y_train)


predictions = model.predict(X_test)

mae = mean_absolute_error(Y_test, predictions)

print("\nRESULTS")
print("MAE:", mae)
print("Sample Predictions:", predictions[:10])

joblib.dump(model, "thickness_model_svm_ms.pkl")
joblib.dump(scaler, "scaler_svm_ms.pkl")

print("\nModel saved successfully.")
```
