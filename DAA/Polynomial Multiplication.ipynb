# Polynomial Multiplication

### Brute Force Approch

```python
def polynomial_multiplication(A,B):
    c = [0]*(len(A) + len(B) - 1)
    for i in range(len(A)):
        for j in range(len(B)):
            c[i + j] += (A[i]*B[j])
    return c
print(polynomial_multiplication([1,2,3,4,5],[5,6,7,8,9]))
```
